```
# !pip install medpy

import medpy as mp
from medpy.io import load

import numpy as np
import matplotlib.pyplot as plt

def plot_mha(path, slice, axis=0):
  image_data, image_head = load(path)
  if axis == 0:
    frame1 = image_data[slice]
  elif axis == 1:
    frame1 = image_data[:, slice, :]
  elif axis == 2:
    frame1 = image_data[:, :, slice]
  else:
    return 'Invalid axis'
  # print(frame1.shape)
  plt.imshow(frame1, cmap='gray')

  
fn1 = '/content/drive/My Drive/Colab Notebooks/siamese_paper/BRATS_HG0001_FLAIR.mha'
fn2 = '/content/drive/My Drive/Colab Notebooks/siamese_paper/BRATS_HG0001_T1.mha'
fn3 = '/content/drive/My Drive/Colab Notebooks/siamese_paper/BRATS_HG0001_T1C.mha'
fn4 = '/content/drive/My Drive/Colab Notebooks/siamese_paper/BRATS_HG0001_T2.mha'
```

```
SLICE = 100
AXIS  = 1

plt.figure(figsize=(10, 10))

for plot, f in zip(range(4), [fn1, fn2, fn3, fn4]):
  plt.subplot(2, 2, plot+1)
  plot_mha(f, SLICE, AXIS)
```
